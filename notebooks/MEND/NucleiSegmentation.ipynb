# Segmentation with WaterShed

```python
sample_img = cv.imread(os.path.join(d_path, 'High squamous intra-epithelial lesion', 'HSIL_1 (1).jpg'))
gray = cv.cvtColor(sample_img, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray, 0, 244, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)
plt.figure(figsize=(10, 10))
plt.imshow(gray, cmap='gray'); plt.imshow(thresh, cmap='gray'); plt.show()
```
